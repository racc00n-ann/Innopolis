## 1. Объясните, как вычисляется качество модели с задачей классификации: назовите основные и вторичные метрики, приведите пример.

Для оценки качества моделей классификации используются различные метрики, которые можно разделить на основные и вторичные в зависимости от задач и специфики использования. Основные метрики обычно применяются в первую очередь, а вторичные — для более детальной оценки.

### Основные метрики классификации:

**Accuracy (достоверность/аккуратность/точность)** — доля правильных ответов модели среди всех ответов. Правильные ответы — это истинно положительные (True Positive) и истинно отрицательные ответы (True Negative).

Вычисляется как отношение числа правильных предсказаний ко всему количеству предсказаний.

$$
\begin{align}
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
\end{align}
$$

где:
TP — True Positive (истинно положительные),

TN — True Negative (истинно отрицательные),

FP — False Positive (ложноположительные),

FN — False Negative (ложноотрицательные).


**Precision (точность), или PPV (Positive Predictive Value)** — это доля верных положительных предсказаний среди всех предсказанных положительных классов.

$$
\begin{align}
Precision = \frac{TP}{TP + FP}
\end{align}
$$


**Recall (полнота), или TPR (True Positive Rate)** — это доля верных положительных предсказаний среди всех фактических положительных классов.

$$
\begin{align}
Recall = \frac{TP}{TP + FN}
\end{align}
$$

Метрика изменяется от 0 до 1.
Интерпретация: способность модели обнаруживать класс 1 вообще, то есть охват класса 1. Метрика зависит от количества ложноотрицательных срабатываний. То есть чем ближе recall к 1, тем меньше вероятность модели допустить ошибку II рода.



**Fb(F-мера)** — это взвешенное среднее гармоническое между precision и recall:

$$
\begin{align}
Recall = (1+β^2)×\frac{Precision × Recall}{β^2 × Precision + Recall}
\end{align}
$$



### Пример расчета метрик:

Предположим, у нас есть 100 наблюдений, и модель дала следующие результаты:

TP=40 (верно предсказаны положительные),

TN=30 (верно предсказаны отрицательные),

FP=20 (ложноположительные),

FN=10 (ложноотрицательные).

Тогда:

$$
\begin{align}
Accuracy &= \frac{40 + 30}{100} = 0.7\ или\ 70\% \\
Precision &= \frac{40}{40 + 20} = 0.67\ или\ 67\% \\
Recall &= \frac{40}{40 + 10} = 0.8\ или\ 80\% \\
F1-score &= 2×\frac{0.67 × 0.8}{0.67 + 0.8} = 0.73\ или\ 73\% \\
\end{align}
$$


 ### Вторичные метрики классификации:

**F1-мера** - частый случай F-меры. Когда β=1 , мы получаем равный вклад для precision и recall, а формула будет выражать простое среднее гармоническое, или метрику F1:

$$
\begin{align}
Recall = 2×\frac{Precision × Recall}{Precision + Recall}
\end{align}
$$

Применяется, когда важно учитывать как ложноположительные, так и ложноотрицательные предсказания.

Метрика равна своему максимуму (1), если и precision, и recall равны 1 (то есть когда отсутствуют как ложноположительные, так и ложноотрицательные срабатывания). Но если хотя бы одна из метрик будет близка к 0, то F1 и будет близка к 0.


**ROC-кривая** — это график, который показывает зависимость между чувствительностью (Recall) и специфичностью (1 — False Positive Rate) при различных порогах классификации.

**AUC-ROC (Area Under the ROC Curve)** - площадь под кривой ROC, которая строится по параметрам TPR (True Positive Rate) и FPR (False Positive Rate).
R):

$$
\begin{align}
TPR = \frac{TP}{TP + FN} \\
FPR = \frac{FP}{FP + TN}
\end{align}
$$

Площадь под ROC-кривой (AUC-ROC) является количественной характеристикой качества бинарной классификации и может принимать значения от 0 до 1. Чем ближе к 1, тем лучше работает модель.



**PR-AUC (Area Under the Precision-Recall Curve)** - площадь под кривой precision-recall. Особенно полезна для сильно несбалансированных данных.

**Matthews Correlation Coefficient (MCC)** - метрика корреляции между предсказанными и реальными классами, принимает значения от -1 до 1.

Коэффициент корреляции Мэтьюса используется в машинном обучении как мера качества бинарных и многоклассовых классификаций . Он учитывает истинные и ложные положительные и отрицательные результаты и обычно рассматривается как сбалансированная мера, которую можно использовать, даже если классы сильно различаются по размеру.

$$
\begin{align}
MCC = \frac{TP⋅TN−FP⋅FN}{\sqrt[]{(TP+FP)(TP+FN)(TN+FP)(TN+FN)}}
\end{align}
$$


## 2. Что такое тестовый набор данных (test set) и для чего он нужен?

**Тестовый набор данных (test set)** — это часть исходного набора данных для тестирования моделм и оценки качества обученной модели.

**Для чего нужен?**
1. Для проверки того, насколько хорошо модель справляется с новыми, незнакомыми данными.
2. Тестовый набор помогает проверить, как модель будет работать на новых данных в реальных условиях.
3. Для проверки итоговой версии модели, чтобы сравнить её с другими и оценить финальное качество.
4. Для избежания завышенной оценки, так как модель проверяется на данных, которых она раньше не видела.


## 3. В чем разница между параметром модели и гиперпараметром алгоритма обучения? Приведите примеры.


Различие между гиперпараметрами и параметрами модели заключается в том, что **гиперпараметры** задаются вручную до начала обучения и определяют характеристики всего процесса обучения, в то время как **параметры модели** вычисляются в процессе обучения на основе данных и оптимизируются для достижения наилучшей производительности.

**Гиперпараметры** можно сравнить с настройками инструмента, с помощью которого вы создаете модель, а **параметры модели** - это результат работы этого инструмента на конкретных данных.

**Гиперпараметры** — это настройки модели, которые определяют её общую структуру и способ обучения. Эти параметры устанавливаются до начала процесса обучения и не изменяются в процессе обучения модели. Гиперпараметры оказывают влияние на то, как модель будет обучаться, какие признаки будут учитываться, и какие ограничения будут наложены на процесс обучения.
**Примерами гиперпараметров** могут быть количество слоев и нейронов в нейронной сети, скорость обучения, шаг градиентного спуска, коэффициент регуляризации и т.д.

**Параметры модели** являются внутренними весами или коэффициентами, которые модель обучает в процессе обучения на основе обучающих данных. Эти параметры изменяются в процессе обучения с целью минимизации функции потерь и достижения наилучшего соответствия между прогнозами модели и реальными значениями целевой переменной.
**Примерами параметров модели** могут быть коэффициенты в линейной регрессии,весовые матрицы и смещения в нейронных сетях, граничные вероятности в логистической регрессии.

# 4. Тренировка моделей и их предсказания

```python
# Подготовка данных для обучения
X = train[['Sex', 'Fare','Pclass', 'Embarked_C', 'Embarked_S']]
y = train['Survived']

# Подготовка тестовых данных
X_test = test[['Sex', 'Fare', 'Pclass', 'Embarked_C', 'Embarked_S']]
```

```python
from sklearn.metrics import accuracy_score

# Функция для обучения модели и сохранения предсказаний
def train_and_save_predictions(model, model_name):
    # Обучение модели
    model.fit(X, y)

    # Предсказание на тестовом наборе
    predictions = model.predict(X_test)

    # Вывод оценки качества модели
    accuracy = accuracy_score(y, model.predict(X))
    print(f'Accuracy of {model_name}: {accuracy:.2f}')

    # Сохранение результатов
    submission = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': predictions})
    submission['Survived'] = submission['Survived'].astype(int)
    filename = f'/content/drive/MyDrive/ДЗ/Titanic_{model_name}.csv'
    submission.to_csv(filename, index=False)
    print(f'Saved: {filename}')
```

Для предсказаний воспользуемся моделями логистической регрессии, древа решений и случайного леса.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Создание и сохранение результатов для каждой модели
log_reg_model = LogisticRegression(max_iter=1000)
train_and_save_predictions(log_reg_model, 'LogisticRegression')

tree_model = DecisionTreeClassifier(random_state=42, max_depth=5, min_samples_split=10)
train_and_save_predictions(tree_model, 'DecisionTree')

rf_model = RandomForestClassifier(random_state=42, n_estimators=100, max_depth=5, max_features='sqrt')
train_and_save_predictions(rf_model, 'RandomForest')
```

 # 5. Предсказание + итоги

Случайный лес показывает наибольшую точность (0.84), что делает его наилучшей моделью среди трех для данной задачи.

```python
#  Подоберем наилучшие параметры модели случайного леса для повышения точности предсказаний.

from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'n_estimators': [100, 200, 500],
    'max_depth': [None, 10, 15, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
}

rf = RandomForestClassifier(random_state=42)

rf_random = RandomizedSearchCV(estimator=rf, param_distributions=param_dist,
                               n_iter=50, cv=3, n_jobs=-1, verbose=2,
                               random_state=42)

rf_random.fit(X, y)

best_params = rf_random.best_params_
print("Best parameters found: ", best_params)
```

```python
# Определение модели с лучшими параметрами
best_rf_model = RandomForestClassifier(
    n_estimators=500,
    min_samples_split=10,
    min_samples_leaf=1,
    max_features='log2',
    max_depth=10,
    random_state=42
)

# Использование функции для обучения и сохранения предсказаний
train_and_save_predictions(best_rf_model, 'Best_RF')
```
